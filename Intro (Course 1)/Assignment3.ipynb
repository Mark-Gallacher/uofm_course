```python
# load in data about energy supply
Energy = pd.read_excel(excel1, 0, # Sheet 1
                    skiprows = 17, skipfooter = 38, # Specify Number of Rows to remove at top and bottom
                    usecols = ('C,D,E,F'), # Specify Cols
                    names=['Country', 'Energy Supply',
                           'Energy Supply per Capita', '% Renewable'], # Rename Col - must match usecols 
                    na_values = "..."  # Specify what NaN should replace in the df
                   ) 
# Changing petajoules into gigajoules
Energy['Energy Supply'] = Energy['Energy Supply'].apply(lambda x: x*1000000)

# Rename Countries and tidy up their names
Energy['Country'] = Energy['Country'].str.replace(r" \(.*\)", "")
Energy['Country'] = Energy['Country'].str.replace(r"\d", "")

Energy['Country'] = Energy['Country'].replace({"Republic of Korea": "South Korea","United States of America":"United States",
                                               "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
                                               "China, Hong Kong Special Administrative Region": "Hong Kong"}) 
                                            #Old : New names in Dict, rename() renames cols not values, so use replace()
```

```python
# Load in GDP data
GDP = pd.read_csv(".\datasets\world_bank.csv", 
                 skiprows = 4, 
                 usecols = ['Country Name','2006','2007','2008','2009',
                            '2010','2011','2012','2013','2014','2015'])

# Renaming some countries
GDP['Country Name']= GDP['Country Name'].replace({"Korea, Rep.": "South Korea",
                                             "Iran, Islamic Rep.": "Iran",
                                             "Hong Kong SAR, China": "Hong Kong"}, 
                                           )
# Renaming column
GDP.rename(columns = {'Country Name': 'Country'}, inplace = True)
GDP.head()
```
